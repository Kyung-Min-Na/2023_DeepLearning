```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# p1 - p2 distance function
def distance(point, p):
    return ((point[0] - p[0]) ** 2 + (point[1] - p[1]) ** 2) ** 0.5
```

```python
# dot function
def draw(group1, group2, c_1, c_2):
    plt.figure(figsize=(8, 8))

    # 산점도 그리기
    group1_x, group1_y = zip(*group1) # 첫번 째 그룹 x, y좌표 추출
    group2_x, group2_y = zip(*group2) # 두번 째 그룹 x, y좌표 추출
    plt.scatter(group1_x, group1_y, color=c_1, s=10, label=f'Group 1 ({len(group1)} points)')  # p1 nearly dot for blue
    plt.scatter(group2_x, group2_y, color=c_2, s=10, label=f'Group 2 ({len(group2)} points)')  # p2 nearly dot for red

    # 그래프 디자인
    plt.title('400 Random Points with Seed 70') # 제목설정
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.xlim(0, 400)
    plt.ylim(0, 400)
    plt.grid(True) # grid 표시
    plt.legend(loc='upper left') # 범례 표시, 왼쪽 위
    return plt
```

# STEP 1 | 랜덤 데이터 생성

```python
# random seed를 70로 지정
random.seed(71)

# 중복없는 400개의 무작위 점 생성
points = set()
while len(points) < 400:
    x = random.randint(0, 400)
    y = random.randint(0, 400)
    point = (x, y)
    if point not in points:
        points.add(point)

# x, y 좌표 추출
x_coords, y_coords = zip(*points)

# 시각화
plt.figure(figsize=(8, 8))
plt.scatter(x_coords, y_coords, color='blue', s=10)  # s = dot size.
plt.title('400 Random Points with Seed 70')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 400)
plt.ylim(0, 400)
plt.grid(True)
plt.show()
```
