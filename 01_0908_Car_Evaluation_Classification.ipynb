```python
import seaborn as sns
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#data loader
with open('../Dataset/car_evaluation/car.data', 'r', encoding='utf-8') as file:
    data = file.readlines()
```

```python
# data frame setup
data_list = []
for line in data:
    columns = line.strip().split(',')
    data_list.append(columns)
    
column_names = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "Label"]
df = pd.DataFrame(data_list, columns=column_names)

df.head()
```

```python
# label encording
label=LabelEncoder()
df['buying_L']=label.fit_transform(df['buying'])
df['maint_L']=label.fit_transform(df['maint'])
df['doors_L']=label.fit_transform(df['doors'])
df['persons_L']=label.fit_transform(df['persons'])
df['lug_boot_L']=label.fit_transform(df['lug_boot'])
df['safety_L']=label.fit_transform(df['safety'])
df["Label_L"]=label.fit_transform(df["Label"])
print(df)

df.head()
```

```python
x = df
x = x.drop(columns=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'Label', 'Label_L'], axis=1)
```

```python
y = df['Label_L']
x
```

```python
# Train test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# SVM

```python
clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

pred_svm = clf_svm.predict(x_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

# LR

```python
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

# DT

```python
clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

# RT

```python
print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)
pred_rf = rf_clf.predict(x_test)
print(accuracy_score(y_test, pred_rf))
print (confusion_matrix(y_test, pred_rf))
```
